#### Build ResNet-20 model for CIFAR-10 datasets

```python
from resnet import build_resnet
model=build_resnet(n=3,input_shape=(32,32,3))
```

```python
model.summary()
```

```python
from resnet import get_input
```

#### Process the input

```python
train_img,train_labels,label_names,test_img,test_labels=get_input('cifar-10-batches-py/')
```

```python
print(train_img.shape)
```

```python
print(test_img.shape)
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
plt.imshow(train_img[1])
plt.show()
print(label_names[train_labels[1]])
```

```python
%matplotlib inline
plt.imshow(test_img[1])
plt.show()
print(label_names[test_labels[1]])
```

```python
from resnet import process_input
```

```python
x_train,x_test,y_train,y_test=process_input(train_img,test_img,train_labels,test_labels)
```

```python
print(x_train.shape)
print(x_test.shape)
```

```python
print(np.max(x_train[0]))
print(np.min(x_train[0]))
```

```python
print(np.max(x_test[0]))
print(np.min(x_test[0]))
```

```python
from resnet import compile_and_fit
```

```python
print(y_train.shape)
```

```python
hist,model=compile_and_fit(x_train,y_train,model,x_test,y_test,batch_size=8,epochs=28)
```

```python
model.evaluate(x=x_train,y=y_train,batch_size=8)
```

```python
model.evaluate(x=x_test,y=y_test,batch_size=8)
```

```python
print(hist.history.keys())
```

```python
print(hist.history['loss'])
```

```python
with open('result.txt','w') as f:
  for value in hist.history['loss']:f.write(str(value)+' ')
  f.write('\n')
  for value in hist.history['accuracy']:f.write(str(value)+' ')
  f.write('\n')
  for value in hist.history['val_loss']:f.write(str(value)+' ')
  f.write('\n')
  for value in hist.history['val_accuracy']:f.write(str(value)+' ')
```

#### Build the Resnet-38 model for CIFAR-10

```python
model_38=build_resnet(n=6,input_shape=(32,32,3))
```

```python
model_38.summary()
```

```python
hist_38,model_38=compile_and_fit(x_train,y_train,model_38,x_test,y_test,batch_size=8,epochs=28)
```

```python
model_38.evaluate(x=x_train,y=y_train,batch_size=8)
```

```python
model_38.evaluate(x=x_test,y=y_test,batch_size=8)
```

```python
x_val=np.arange(1,29,1)
```

```python
%matplotlib inline
plt.plot(x_val,hist.history['loss'],label='loss_20')
plt.plot(x_val,hist.history['val_loss'],label='val_loss_20')
plt.plot(x_val,hist_38.history['loss'],label='loss_38')
plt.plot(x_val,hist_38.history['val_loss'],label='val_loss_38')
plt.xlabel('Epoch')
plt.ylabel('loss')
plt.legend(loc='upper right')
plt.savefig('loss_diagram.png')
plt.show()
```

```python
%matplotlib inline
plt.plot(hist.history['accuracy'],label='acc_20')
plt.plot(hist.history['val_accuracy'],label='val_acc_20')
plt.plot(hist_38.history['accuracy'],label='acc_38')
plt.plot(hist_38.history['val_accuracy'],label='val_acc_38')
plt.xlabel('Epoch')
plt.ylabel('accuracy')
plt.legend(loc='upper left')
plt.savefig('accuracy_diagram.png')
plt.show()
```
