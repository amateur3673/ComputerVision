## Selective Search Segmentation algorithm

Note the keypoint in the concept of the **Selective Search** algorithm.

```python
import numpy as np
import cv2
import source
import matplotlib.pyplot as plt
```

```python
img=cv2.imread('Images/Figure1.jpg')
img=img[:,:,::-1]
%matplotlib inline
plt.imshow(img)
plt.show()
```

```python
img_comp,list_comp=source.retrieve_components(img)
```

```python
print('Number of components: {}'.format(len(list_comp)))
```

```python
segmented_img=source.initial_segment(list_comp,(img.shape[0],img.shape[1]))
```

```python
segmented_img=segmented_img[:,:,::-1]
%matplotlib inline
plt.imshow(segmented_img)
plt.show()
```

```python
neighbor_mat=source.find_neighbors(img_comp,len(list_comp))
```

```python
list_comp,neighbor_mat=source.initial_merge(list_comp,neighbor_mat)
```

```python
print('Number of components after conducting the initial merge: {}'.format(len(neighbor_mat)))
```

```python
segmented_img=source.initial_segment(list_comp,(img.shape[0],img.shape[1]))
```

```python
%matplotlib inline
plt.imshow(segmented_img)
plt.show()
```
