```
import tensorflow as tf
import numpy as np
```

```
from model import build_model
```

```
g,d,gan=build_model(n_input=100)
```

```
(X_train,y_train),(X_test,y_test)=tf.keras.datasets.mnist.load_data()
X_train=X_train.reshape(X_train.shape[0],784)
X_train=X_train.astype(np.float32)/255.0
```

```
from model import train_model
```

```
losses,(g,d,gan)=train_model(X_train,(g,d,gan),epochs=100)
```

```
import matplotlib.pyplot as plt
```

```
epochs=np.arange(1,len(losses['D'])+1)
%matplotlib inline
plt.plot(epochs,losses['G'],label='Generator loss')
plt.plot(epochs,losses['D'],label='Discriminator loss')
plt.legend(loc='upper right')
plt.title('GAN_mnist Loss')
plt.xlabel('Epoch')
plt.ylabel('loss')
plt.savefig('mnist_loss.png')
plt.show()
```

```
g.save_weights('mnist_generator.h5')
```

```
d.save_weights('mnist_dis.h5')
```
